## Plotting a 3D model

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import measure, morphology
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
```

```python
def plot_3d(image, threshold=-300):
    """Plots the image as a 3D model.
    """
    # Position the scan upright,
    # so the head of the patient would be at the top facing the camera
    p = image.transpose(2, 1, 0)
    print("Transpose finished")

    verts, faces, _, _ = measure.marching_cubes(p, threshold)
    print("Marching cube finished")
    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_subplot(111, projection='3d')
    print("Subplot finished")

    # Fancy indexing: `verts[faces]` to generate a collection of triangles
    mesh = Poly3DCollection(verts[faces], alpha=0.70)
    print("Poly3DCollection finished")
    face_color = [0.45, 0.45, 0.75]
    mesh.set_facecolor(face_color)
    print("facecolor finished")
    ax.add_collection3d(mesh)
    print("Add collection finished")

    ax.set_xlim(0, p.shape[0])
    ax.set_ylim(0, p.shape[1])
    ax.set_zlim(0, p.shape[2])
    print("Now drawing")
    
    plt.show()
```

```python
# Drawing is very slow (~3 minutes), but it works. We should find a faster way though.

image_51508704 = np.load('/home/shared/data/data-20180405/patient-ZVRO2HKWM7JSS6IA.npy')
print("Hi")
plot_3d(image_51508704, 400)
```
