```python
import os
import typing
import numpy as np
import pathlib
```

```python
def load_data(data_dir: str) -> typing.Dict[str, np.ndarray]:
    """Returns a dictionary which maps patient ids
    to patient pixel data."""
    data_dict = {}
    for filename in os.listdir(data_dir):
        patient_id = filename[:-4] # remove .npy extension
        data_dict[patient_id] = np.load(pathlib.Path(data_dir) / filename)
    return data_dict
```

```python
train = load_data('/home/harold_triedman/elvo-analysis/data/preds/train')
val = load_data('/home/harold_triedman/elvo-analysis/data/preds/val')
test = load_data('/home/harold_triedman/elvo-analysis/data/preds/test')
```

```python
from scipy.ndimage.interpolation import zoom
import pandas as pd
```

```python
def clean_up(data):
    new_data = {}
    to_return_labels = {}
    labels = pd.read_csv('/home/harold_triedman/elvo-analysis/binary_vectors.csv')
    labels = labels.set_index('Unnamed: 0')
    for id_, arr in list(data.items()):
        if arr.dtype == 'O':
            continue
        try:
            print(labels.loc[id_]['labels'])
        except:
            continue
        arr = zoom(arr, (800 / len(arr)))
        new_data[id_] = arr
        to_return_labels[id_] = int(labels.loc[id_]['labels'])
    return new_data, to_return_labels
```

```python
train, train_labels = clean_up(train)
val, val_labels = clean_up(val)
test, test_labels = clean_up(test)
```

```python
print(len(train), len(train_labels))
print(len(val), len(val_labels))
print(len(test), len(test_labels))
```

```python
import random
```

```python
def get_usable_data(data, labels):
    x = []
    y = []
    for id_, arr in list(data.items()):
        x.append(arr)
        y.append(labels[id_])
    tmp = list(zip(x, y))
    random.shuffle(tmp)
    x, y = zip(*tmp)
    return np.asarray(x), np.asarray(y)
```

```python
x_train, y_train = get_usable_data(train, train_labels)
x_val, y_val = get_usable_data(val, val_labels)
x_test, y_test = get_usable_data(test, test_labels)
```

```python
print(x_train.shape, y_train.shape)
print(x_val.shape, y_val.shape)
print(x_test.shape, y_test.shape)
```

```python
x_train = np.append(x_train, x_val, axis=0)
y_train = np.append(y_train, y_val, axis=0)
```

```python
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1]))
y_train = np.reshape(y_train, (y_train.shape[0], 1))
print(x_train.shape)
print(y_train.shape)
```

```python
import autosklearn.classification
```

```python
cls = autosklearn.classification.AutoSklearnClassifier()
cls.fit(x_train, y_train)
```

```python
y_hat = cls.predict(x_test)
```

```python
import sklearn.metrics
```

```python
print("Accuracy score", sklearn.metrics.accuracy_score(y_test, y_hat))
print("Confusion Matrix", sklearn.metrics.confusion_matrix(y_test, y_hat))
```
