```python
import os
import pathlib
import typing
from scipy.ndimage.interpolation import zoom
import pandas as pd
import random
import numpy as np
import autosklearn.classification
from sklearn import model_selection, metrics
```

```python
def load_data(data_dir: str) -> typing.Dict[str, np.ndarray]:
    """Returns a dictionary which maps patient ids
    to patient pixel data."""
    data_dict = {}
    for filename in os.listdir(data_dir):
        patient_id = filename[:-4] # remove .npy extension
        data_dict[patient_id] = np.load(pathlib.Path(data_dir) / filename)
    return data_dict
```

```python
all_data_dict = load_data('/home/harold_triedman/elvo-analysis/data/normal/axial/')
```

```python
train_labels_df = pd.read_csv('/home/harold_triedman/elvo-analysis/data/preprocessed/train.csv')
val_labels_df = pd.read_csv('/home/harold_triedman/elvo-analysis/data/preprocessed/val.csv')
```

```python
print(len(train_labels_df))
```

```python
def process_images(data: typing.Dict[str, np.ndarray]):
    return {id_: arr for id_, arr in data.items() if len(arr) != 1} # Remove the bad image

def process_pt2(data: typing.Dict[str, np.ndarray], labels: pd.DataFrame):
    return_data = {}
    for id_, arr in data.items():
        if id_ in labels.loc[:, '0'].tolist():
            return_data[id_] = arr
    return return_data
```

```python
all_data_dict = process_images(all_data_dict)
train_data = process_pt2(all_data_dict, train_labels_df)
val_data = process_pt2(all_data_dict, val_labels_df)
```

```python
def process_labels(labels: pd.DataFrame, data: typing.Dict[str, np.ndarray]):
    # TODO: Remove duplicate HLXOSVDF27JWNCMJ, IYDXJTFVWJEX36DO from ELVO_key
    labels = labels.loc[~labels.index.duplicated()] # Remove duplicate ids
    labels = labels.loc[list(data.keys())]
    assert len(labels) == len(data)
    return labels
```

```python
def interpolate(image: np.ndarray):
#     image = np.transpose(image, (1, 2, 0))
    dims = np.shape(image)
    print(dims)
    image = zoom(image, (220 / dims[0],
                         220 / dims[1]))
    return image

                 
def process_data(data: typing.Dict[str, np.ndarray]) -> typing.Dict[str, np.ndarray]:
    processed = {}
    for id_, arr in data.items():  
        print(f'processing scan {id_}')
        processed[id_] = interpolate(arr)
        print(processed[id_].shape)
    return processed
```

```python
processed_data_train = process_data(train_data)
processed_data_val = process_data(val_data)
```
